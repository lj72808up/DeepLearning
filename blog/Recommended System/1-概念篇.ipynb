### 一. 什么是推荐系统
1. 它能做什么  
推荐系统可以把那些最终会在用户（User）和物品（ItemItem）之间产生的连接提前找出来
2. 它需要做什么  
推荐系统需要已经存在的连接，从已有的连接去预测未来的连接。
3. 它怎么做  
  1. 机器推荐  
  2. 网红推荐 : 为用户找到适合他的网红

### 二. 推荐系统的问题模式
因为推荐系统的使命是为用户和物品建立连接, 建立方式是提前找出那些潜在连接呈现出来, 是一个预测问题.   
所以,从预测角度出发, 推荐系统分为两种模式: `评分预测`和`行为预测`  
#### 1. 评分预测  
1. 评分预测是说:  
   假如用户消费完物品后, 会给出一个评分. 于是我们在用户还没消费的物品上, 预测用户可能给出的评分,并按照评分大小推荐给用户  
2. 评分模型的拼盘标准:  
   首先评分模型会在用户历史的消费物品上预测得分, 然后模型给估计这些历史物品的评分, 并使用均方根误差$RMSE$优化模型.  
   这个思路是传统的回归问题$$RMSE=\sqrt { \frac { \sum _{ t=1 }^{ n }{ { \left( \widehat { { y }_{ t } } -{ y }_{ t } \right)  }^{ 2 } }  }{ n }  } $$
3. 评分预测的问题  
 1. 数据不易收集  
  评分模型需要用户完整的完成所有消费流程, 并最终给出评分, 而用户往往无法完成整个流程  
 2. 数据质量无法保证  
  用户的历史评分可能是随便评价的  
 3. 评分分布不稳定  
  用户在不同时期的标准不同, 评分差异较大, 而且人和人的评分标准差距很大

#### 2. 行为预测
1. 用户评分是用户的显式反馈, 还有一类返回是隐式反馈, 用常识用户的各类型为, 也就是另一类预测: 行为预测  
2. 行为预测的方式  
  1. 直接预测行为发生的概率(CTR点击率预估)  
  2. 预测物品的相对排序
3. 行为预测的特点  
  1. 隐式反馈的数据比显式反馈的数据更加稠密.  
   毕竟评分数据是很稀疏的 
  2. 隐式反馈更代表用户的真实想法  
   比如,假设你并不同意Trump的观点,于是你给了很低的评分.但是你还是很像看到他, 以便吐槽他  
  3. 隐式反馈常常和目标函数挂钩, 更能在AB测试中通过
   
#### 3. 推荐系统的潜在问题 
1. 冷启动问题
 推荐系统是数据贪婪型应用，所谓数据贪婪型应用，就是对数据的需求永无止境. 冷启动问题就是数据匮乏问题
2. 探索与利用问题  
 探索指无视用户兴趣, 推荐一些历史行为中体现不出来的物品. 利用是指基于用户历史数据给出的推荐物品. 这之间的比例就是探索与利用问题(explore&exploit)
3. 安全问题  
 推荐系统被攻击后, 可能会产生以下不良后果  
  1. 推荐不靠谱的产品
  2. 收集了不靠谱的脏数据(这种影响会一直持续在系统中, 很难完全消除)  
  3. 损失经济利益
