# 1. 时间序列简介
----------------
时间序列是数理统计的一个分支.统计分为回归与时序  
* 回归 : 处理截面数据.同类型无时序数据.要求feature:  
( 1 )不能自相关: 同个属性自身没有相关性 (同是人口属性,但不同城市之间人口数量不能有关联)  
( 2 )没有多重共线性: 多个feature之间没有线性关系  
( 3 )不存在异方差: 回归方程的残差项,需要满足同样的方差. 例如线性回归,就是假设残差项${(y-a)}^2$服从正态分布下推导而来的     
* 时序 : 研究自相关

### 一. 符号表示
#### 1. 时间序列的定义  
( 1 ) 随机序列: 抽象概念, 按时间顺序排列的一组随机变量. 用${ X }_{ 1 },{ X }_{ 2 }..{ X }_{ t }$表示   
( 2 ) 观测值序列: 随机序列得到的n个有序观测数据,用${ x }_{ 1 },{ x }_{ 2 }..{ x }_{ n }$表示   
( 3 ) 我们的目标是通过对观测值序列的学习, 得到抽象的随机序列表达式

### 二. 时间序列分析方法
#### 1. 描述性分析
通过对数据的绘图观测, 找到序列的发展规律, 是进行统计时需分析的第一步
#### 2. 统计时序分析
1. 频域分析:  
( 1 ) 任何一种无趋势时间序列都可以分解为若干"不同频率的周期波动"   
( 2 ) 主要是对观测序列形成的函数图像, 分解为多个函数的和的形式. 每个函数都是不同频率的周期函数  
2. 时域分析  
( 1 ) 利用统计分析描述序列值之间的相关关系, 拟合出统计模型    
( 2 ) 时需分析的步骤:  
 * 考察观测值特征
 * 拟合模型
 * 确定模型口径
 * 检验模型,优化模型
 * 利用拟合好的模型推断统计性质,预测序列发展

# 2. 时间序列可解决的数据特点
-------------------
### 一. 平稳性
#### 1. $t$ 时刻的统计特征量
1. 均值: ${ \mu  }_{ t }=E{ X }_{ t }=\int _{ -\infty  }^{ +\infty  }{ xdF(x) } $   
2. 方差: $D{ X }_{ t }=E{ \left( { X }_{ t }-{ \mu  }_{ t } \right)  }^{ 2 }=\int _{ -\infty  }^{ +\infty  }{ { \left( x-{ \mu  }_{ t } \right)  }^{ 2 } } dF(x)$   
3. 自协方差: $\gamma (t,s)=E\left( { X }_{ t }-{ \mu  }_{ t } \right) \left( { X }_{ s }-{ \mu  }_{ s } \right) $   
4. 自相关系数: $\rho \left( t,s \right) =\frac { \gamma (t,s) }{ \sqrt { D{ X }_{ t }*D{ X }_{ s } }  } $

#### 2. 平稳时间序列
1. 严平稳:   
 序列所有统计量都不会随时间推移而变化  
定义:$\forall 正整数\tau ,且{ t }_{ 1 },{ t }_{ 2 }...{ t }_{ m }\in T,有{ F }_{ { t }_{ 1 },{ t }_{ 2 }...{ t }_{ m } }\left( { x }_{ 1 },{ x }_{ 2 }...{ x }_{ m } \right) ={ F }_{ { t }_{ 1 }+\tau ,{ t }_{ 2 }+\tau ...{ t }_{ m }+\tau  }\left( { x }_{ 1 },{ x }_{ 2 }...{ x }_{ m } \right) $  
2. 宽平稳:   
 使用序列的统计量来判断是否平稳.它认为序列的性质主要由其低阶矩表现. 只要其低阶矩(二阶矩)平稳, 就能保证其序列平稳  
定义: $E{ X }_{ t }=常数\\ \gamma (t,s)=\gamma (k,k+s-t)\quad $

#### 3. 平稳时间序列的统计性质
1. 常数均值
2. 自协方差函数与自相关函数只依赖时间的平移长度,与时间起止点无关   
 延迟为k的协方差函数: $\gamma \left( k \right) =\gamma \left( t,t+k \right) $    
 延迟k的自相关函数: ${ \rho  }_{ k }=\frac { \gamma \left( k \right)  }{ \gamma \left( 0 \right)  } $

#### 4. 平稳性图示
1. 时序图:  
 平稳性时间序列,均值和方差为常数. 因此其时序图表现为序列始终在一个常数值附近随机波动, 且波动范围有界, 无明显趋势及周期性特征  
2. 自相关图:  
 平稳序列通常有短期相关性. 因此, 平稳时间序列的自相关系数图表现为: 随着延迟期数的增加, 平稳序列的自相关系数很快趋于零.即$\gamma \left( t,t+1 \right) \neq 0,但\gamma \left( t,t+k \right) \rightarrow 0$

### 二. 纯随机性

#### 1. 何为纯随机序列
1. 纯随机序列也叫白噪声序列,需要满足:  
$(1)\quad \forall t\in T,\quad E{ X }_{ t }=\mu \\ (2)\quad \gamma \left( t,s \right) =\begin{cases} 0\quad (t\neq s),\quad 表示各序列之间没有任何相关关系,为"没有记忆"的序列 \\ { \sigma  }^{ 2 }\quad (t=s),\quad 马尔科夫定理说明,只有满足方差齐性,由最小二乘法得到的参数估计值才有效 \end{cases}\\ \\ $  
2. 纯随机序列属于平稳序列  
 纯随机序的定义需要满足: (1)常数均值, 常数方差 (2)自协方差函数只与时间平移长度有关,与起止点无关. 因为$\gamma \left( s\neq t \right) =0$  
 因此, 平稳性数据包含了白噪声数据, 纯随机性验证, 就是把平稳性数据中, 不具备研究价值的数据刨除在外

#### 2. Barlett定理 
1. 定义:
 如果一个序列是纯随机序列, 得到观察期数为n的观测序列, 则该序列延迟非零期的样本自相关系数近似服从$N\left( 0,\frac { 1 }{ n }  \right) $  
 即$\widehat { { \rho  }_{ k } } \sim N\left( 0,\frac { 1 }{ n }  \right) $  

#### 3. Q统计量~$X^2(m)$   
1. $Q=n\sum _{ k=1 }^{ m }{ { \widehat { { \rho  }_{ k } }  }^{ 2 } } \sim { X }^{ 2 }\left( m \right) $  
2. 证明:  
$\because D\left( \widehat { { \rho  }_{ k } }  \right) =\frac { 1 }{ n } \\ \therefore D\left( \sqrt { n } \widehat { { \rho  }_{ k } }  \right) ={ (\sqrt { n } ) }^{ 2 }D\left( \widehat { { \rho  }_{ k } }  \right) =n*\frac { 1 }{ n } =1\\ \therefore \sqrt { n } \widehat { { \rho  }_{ k } } \sim N\left( 0,1 \right) \Longrightarrow \sum _{ i=1 }^{ m }{ { \left( \sqrt { n } \widehat { { \rho  }_{ k } }  \right)  }^{ 2 } } \sim { X }^{ 2 }\left( m \right) $  

#### 4. LB统计量~$X^2(m)$  
$LB=n\left( n+2 \right) \sum _{ k=1 }^{ m }{ \left( \frac { { \widehat { { \rho  }_{ k } }  }^{ 2 } }{ n-k }  \right)  } $

#### 5. 假设检验
1. 因为Q统计量和LB统计量都服从$X^2(m)$分布, 所以可以计算Q与LB统计量对应的p值, 若p值小于$\alpha $, 则表示统计量不再置信区间内, 认为该序列不是纯随机序列  
2. 实际应用中, 一般对取延迟6期或12期的值进行统计量p值计算, 若$p\preceq 0.05$,则统计量小于0.95分位点, 认为不是纯随机序列
