# 1. 时间序列简介
----------------
时间序列是数理统计的一个分支.统计分为回归与时序  
* 回归 : 处理截面数据.同类型无时序数据.要求feature:  
( 1 )不能自相关: 同个属性自身没有相关性 (同是人口属性,但不同城市之间人口数量不能有关联)  
( 2 )没有多重共线性: 多个feature之间没有线性关系  
( 3 )不存在异方差: 回归方程的残差项,需要满足同样的方差. 例如线性回归,就是假设残差项${(y-a)}^2$服从正态分布下推导而来的     
* 时序 : 研究自相关

### 一. 符号表示
#### 1. 时间序列的定义  
( 1 ) 随机序列: 抽象概念, 按时间顺序排列的一组随机变量. 用${ X }_{ 1 },{ X }_{ 2 }..{ X }_{ t }$表示   
( 2 ) 观测值序列: 随机序列得到的n个有序观测数据,用${ x }_{ 1 },{ x }_{ 2 }..{ x }_{ n }$表示   
( 3 ) 我们的目标是通过对观测值序列的学习, 得到抽象的随机序列表达式

### 二. 时间序列分析方法
#### 1. 描述性分析
通过对数据的绘图观测, 找到序列的发展规律, 是进行统计时需分析的第一步
#### 2. 统计时序分析
1. 频域分析:  
( 1 ) 任何一种无趋势时间序列都可以分解为若干"不同频率的周期波动"   
( 2 ) 主要是对观测序列形成的函数图像, 分解为多个函数的和的形式. 每个函数都是不同频率的周期函数  
2. 时域分析
( 1 ) 利用统计分析描述序列值之间的相关关系, 拟合出统计模型    
( 2 ) 时需分析的步骤:  
$考察观测值特征\\ 拟合模型\\ 确定模型口径\\ 检验模型,优化模型\\ 利用拟合好的模型推断统计性质,预测序列发展\quad $
